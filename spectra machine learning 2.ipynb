```python
%matplotlib inline
import tandems
```

```python
tandems.generate_spectral_bins(loadFullSpectra = True,
        method='clusters', numFeatures = 200, fname='latitude40')
```

```python
s = tandems.load('lat40/E latitude40.bins 6 6 1000 1517873359')
s.__init__(specsFile='latitude40.clusters.npy')
s.bins =  list(range(1,21))
s.recalculate()
s.bins =  list(range(1,20))+[21,20]
s.plot()
s.Iscs = 0
s.spectra = 0
s.save()
```

```python
s=tandems.effs(junctions=6,cells=180,convergence=True,concentration=1000,name='lat40/E',specsFile='latitude40.clusters.npy')
s.findGaps()
s.plot()
s.save()
```

```python
s=tandems.load('lat40/E latitude40.clusters 6 6 1000 42252375')
s.__init__(name='lat40/E',specsFile='latitude40.bins.npy',convergence=False,bins=list(range(1,21)))
s.recalculate()
s.convergence=True
s.bins =  list(range(1,20))+[21,20]
s.plot()
```

```python
s0=tandems.load('lat40/E latitude40.clusters 6 6 1000 42252375')
s.compare(s0, r1=(0.8,0.1,0,1), r2=(1,0.1,0,0.4), b1=(0,0.1,1,0.6), b2=(0,0.1,1,0.4))
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
plt.figure()
plt.xlabel('Number of spectra')
plt.ylabel('Standard deviation of efficiency overestimate (%)')
plt.xticks(list(range(1,21)))
plt.xlim(1,20) 
plt.tick_params(axis='y', right='on')
ec = 100*( s.effs[ : , 1:-1 ] - s.effs[ : , -1 ][:, None] )
ecm = np.mean( ec, axis=0 )
ece = np.sqrt ( np.mean( ( ec - ecm )**2, axis = 0 ) )
plt.ylim( 0, 1.01*( ece )[0] ) 
plt.plot( range(1,21), ece, color ='b') #  color = LGBT(auxIs[i*s.junctions])
ec = 100*( s0.effs[ : , 1:-1 ] - s0.effs[ : , -1 ][:, None] )
ecm = np.mean( ec, axis=0 )
ece = np.sqrt ( np.mean( ( ec - ecm )**2, axis = 0 ) )
plt.plot( range(1,21), ece, color ='r') #  color = LGBT(auxIs[i*s.junctions])
plt.savefig(s.name+' '+s.specsFile.replace('.npy', '').replace('.','-')+' convergence compare_ '+str(int(s.junctions))+' '+str(int(s.topJunctions))+' '+str(int(s.concentration))+' '+s.autoSuffix, dpi=600)
```
